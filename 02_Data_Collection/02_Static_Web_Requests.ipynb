````python
src = resp.text
````

````python
soup = BeautifulSoup(src, 'lxml')
print(soup)
````

## quiz : 주요뉴스의 목록을 추출해보자

````python
# quiz : 주요뉴스의 목록을 추출해보자
# 뉴스 목록 찾기
srclist = soup.select('.section_strategy a')
srclist
````

````python
url = srclist[0]['href']
PATH + url
````

````python
from urllib.parse import urljoin
urljoin(PATH, url)
````

````python
# quiz : 네이버 증권 주요뉴스의 제목과 url 을 모두 추출하기
# 제목 변수, url 변수 만들기
# 제목과 url 데이터 목록 만들기
news_title = []
news_url = []

for item in srclist:
    title = item.text
    url = urljoin(PATH, item['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)   
````

````python
# 데이터프레임 변환
import pandas as pd
df = pd.DataFrame( {'제목':news_title, '주소':news_url}   )
df
````
